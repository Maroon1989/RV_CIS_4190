```python
import numpy as np
import pandas as pd
from Factor_Calculate import Factor
```

```python
factors = Factor()
```

```python
def calc_wap2(df: pd.DataFrame) -> pd.Series:
    wap = (df['bid_price2'] * df['ask_size2'] + df['ask_price2'] * df['bid_size2']) / (df['bid_size2'] + df['ask_size2'])
    return wap
```

```python
# 计算feature的函数
def wap2(df:pd.DataFrame) -> pd.Series:
    df['wap2'] = calc_wap2(df)
    return df
```

```python
# 传参格式
func_call = {'wap2':[wap2,[np.sum, np.mean, np.std]]}
# 调试用，生成1个feature
df = factors.book_feature(func_call=func_call,stock_id='stock_id=0')
# 对每个股票代码生成1个feature

# df = factors.make_book_feature(func_call)


# data = factors.load_data(book=True,stock_id='stock_id=0')
# data
# factors.make_book_feature(func_call)
```
